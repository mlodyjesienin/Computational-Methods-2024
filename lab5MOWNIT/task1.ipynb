```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

s = np.linspace(0, 2 * np.pi, 1000)
t = np.linspace(0, np.pi, 10)
x = np.outer(np.cos(s), np.sin(t))
y = np.outer(np.sin(s), np.sin(t))
z = np.outer(np.ones(np.size(s)), np.cos(t))

# Poczenie wsprzdnych w jedn macierz
sphere = np.array([x.flatten(), y.flatten(), z.flatten()])
print(sphere.shape)

# Wygenerowanie losowej macierzy 3x3
A = np.random.rand(3, 3)

# Przeksztacenie sfery w elipsoid za pomoc macierzy
elipsoid = np.dot(A, sphere)

# Przeksztacenie wynikowej macierzy z powrotem do pierwotnego ksztatu
x1 = elipsoid[0].reshape(x.shape)
y1 = elipsoid[1].reshape(y.shape)
z1 = elipsoid[2].reshape(z.shape)

# Wywietlenie wizualizacji uzyskanego wyniku w 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1, y1, z1, cmap='viridis')
plt.show()

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, y1, z1, c='b', marker='o')
plt.show()
```
