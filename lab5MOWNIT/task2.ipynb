```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open("lenna.png").convert("L")  # Otwórz obraz w skali szarości
pixels = np.array(image)


U, S, V = np.linalg.svd(pixels, full_matrices=True)
```

```python
k = 40  # Liczba wartości osobliwych do wykorzystania (można dostosować)
```

```python
def outer_product(k,S,U,V):
    result = np.zeros((512,512))
    for i in range(k):
        result += S[i]*np.outer(U[:,i],V[i])
        
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(pixels, cmap='gray')
    plt.title('Oryginalny obraz')

    plt.subplot(1, 2, 2)
    plt.imshow(result, cmap='gray')
    plt.title(f'Kompresja do {k} wartości osobliwych')

    plt.show()
```

```python
def matrix_multiplying(k,S,U,V):
    U = U[:, :k] 
    S = np.diag(S[:k])
    V = V[:k, :]
    compressed_pixels = np.dot(U, np.dot(S, V))
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(pixels, cmap='gray')
    plt.title('Oryginalny obraz')

    plt.subplot(1, 2, 2)
    plt.imshow(compressed_pixels, cmap='gray')
    plt.title(f'Kompresja do {k} wartości osobliwych')

    plt.show()
```

```python
matrix_multiplying(k,S,U,V)
outer_product(k,S,U,V)
```
